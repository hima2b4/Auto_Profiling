製造工程での検査記録等、多くのデータカラムで構成されたデータセットの基本統計量、ヒストグラム、相関係数行列等を表示、可視化するライブラリです。\
基本、データセット（csv）を指定するだけであり、データゼットの全容把握を非常に効率的に行うことができます。

## **データについて**

|data |data|data |…|data |…|data |

|･ |･|･ |…|･ |…|･ |
|･ |･|･ |…|･ |…|･ |

- [**注意**] csvデータは文字コードを「UTF-8」としてください。

**Describe**：列のタイトルからアクセスして、各列の統計分析を見ることができます。
Filterオプションでは、データにフィルターをかけて、特定の条件を満たす行だけを表示することができます。単純なフィルタリング操作であれば、列フィルタリングを利用するのが簡単です。より複雑なフィルタリング操作は、右側に定義されているコードに従って、フィルタリングルールを記述する必要があります。

**Correlations**：変数がお互いにどのように相関しているかを表示します。

**Chart**
利用可能なグラフィック（線、棒、散布図、円、ワードクラウド、ヒートマップ、3D散布図、サーフェス）の中から必要なものを生成できます。
また、各チャートに表示する変数や、表示する最大値や最小値などのパラメータを定義することができます。

**Heatmap**：最も洗練されたオプションの一つがヒートマップです。これは、データテーブルの各セルの値に応じて色を表示するものです。
