製造工程での検査記録等、多くのデータカラムで構成されたデータセットの基本統計量、ヒストグラム、相関係数行列等を表示、可視化するライブラリです。\
基本、データセット（csv）を指定するだけであり、データゼットの全容把握を非常に効率的に行うことができます。

## **データについて**
1. **csvデータの表形式は以下としてください。**

|列1|列2|列3|列4|列5|列6|…|
|---|---|---|---|---|---|---|
|data |data|data |…|data |…|data |
|data |data|data |…|data |…|data |
|･ |･|･ |…|･ |…|･ |
|･ |･|･ |…|･ |…|･ |
|･ |･|･ |…|･ |…|･ |

- [**注意**] csvデータは文字コードを「UTF-8」としてください。

## **実行手順**
1. メニューバーの「ランタイム」から「すべてのセルを実行」をクリック。
2. ライブラリインストール完了後、[**ファイル選択**] を促されるので、クリックしてデータ（csvファイル）を指定する。

```
import pandas as pd
import warnings
from pandas_profiling import ProfileReport
from pandas_profiling.utils.cache import cache_file

warnings.filterwarnings('ignore')
```

## **ファイル指定**


```
#@title csvファイル（UTF-8）を指定してください
from google.colab import files
#print('csvファイル(UTF-8）を指定してください')
uploaded = files.upload()
```

```
#@title 読込みデータ表示（先頭5行）
if len(uploaded.keys()) != 1:
    print("アップロードは１ファイルにのみ限ります")
else:
    target = list(uploaded.keys())[0]

df = pd.read_csv(target,index_col=0)
profile = ProfileReport(df)
df.head()
```

```
#@title Profile結果（html）をローカルファイルに保存
from google.colab import files
filename =  'Profile‗result.html'
profile.to_file(filename) 
files.download(filename)
```
