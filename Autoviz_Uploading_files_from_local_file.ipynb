製造工程での検査記録等、多くのデータカラムで構成されたデータセットの基本統計量、ヒストグラム、相関係数行列等を表示、可視化するライブラリです。\
基本、データセット（csv）を指定するだけであり、データゼットの全容把握を非常に効率的に行うことができます。

## **データについて**
1. **csvデータの表形式は以下としてください。**

|列1|列2|列3|列4|列5|列6|…|
|---|---|---|---|---|---|---|
|data |data|data |…|data |…|data |
|data |data|data |…|data |…|data |

|･ |･|･ |…|･ |…|･ |
|･ |･|･ |…|･ |…|･ |

- [**注意**] csvデータは文字コードを「UTF-8」としてください。
